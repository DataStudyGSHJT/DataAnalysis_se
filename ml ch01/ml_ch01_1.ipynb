## 01-1. 인공지능과 머신러닝, 딥러닝
### 인공지능이란
**Artificial Intelligence**  
: 사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템을 만드는 기술  

  image.png

* 인공지능 태동기  
  1943. 뇌의 뉴런 개념 발표(워런 매캘러, 월터 피츠)  
  1950. 튜링 테스트 발표(앨런 튜링)  
  1956. 다트머스 AI 컨퍼런스  
    
* 인공지능 황금기  
  1957. 퍼셉트론 - 로지스틱 회귀의 초기 버전  
  1959. 시각 피질 뉴런 기능 연구 - 고양이

* 인공일반지능(또는 강인공지능)  
 : 사람과 구분하기 어려운 지능을 가진 컴퓨터 시스템
* 약인공지능  
 : 특정 분야에서 사람의 일을 보조하는 역할만 가능

### 머신러닝이란
**Machine Learning**  
: 규칙을 일일이 프로그래밍하지 않아도 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야

* 사이킷런 scikit-learn  
 : 파이썬 API를 사용하는 머신러닝 라이브러리
 - 안정적
 - 검증된 성능

### 딥러닝이란
**Deep Learning**  
: 많은 머신러닝 알고리즘 중 인공 신경망을 기반으로 한 방법들의 통칭

1998. LeNet-5 - 최초의 합성곱 신경망: 손글씨 숫자 인식 (얀 르쿤)
2012. AlexNet - 이미지 분류 (제프리 힌턴)
2015. 텐서플로 Tensorflow - 오픈소스 딥러닝 라이브러리 (구글)
2016. 알파고
2018. 파이토치 Pytorch - 오픈소스 딥러닝 라이브러리 (페이스북)
