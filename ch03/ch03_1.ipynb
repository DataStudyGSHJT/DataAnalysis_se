## 중복된 행 찾기

```python
sum(ns_book.duplicated()) # duplicated() : 데이터프레임에 있는 모든 열을 기준으로 중복된 행을 찾음
```

```python
sum(ns_book.duplicated(subset=['도서명', '저자', 'ISBN'])) # subset=[] : 기준이 되는 열 설정
```

```python
# drop_duplicates() : 중복된 행 삭제
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False) # keep=False : 중복된 행을 모두 True로 표시한 boolean array 반환
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

## 그룹별로 모으기

```python
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
count_df
```

```python
# group_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False)
# loan_count = group_df.sum()
loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()
loan_count.head()
```

## 원본 데이터 업데이트하기

```python
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
unique_rows = ~dup_rows
ns_book3 = ns_book[unique_rows].copy()
```

```python
sum(ns_book3.duplicated(subset=['도서명', '저자', 'ISBN', '권']))
```

### 원본 데이터프레임 인덱스 설정하기

```python
ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
ns_book3.head()
```

### 업데이트하기: update() 메서드

```python
ns_book3.update(loan_count) # update() : 다른 데이터프레임을 사용해 원본 데이터프레임의 값을 업데이트하는 경우
ns_book3.head()
```

```python
ns_book4 = ns_book3.reset_index() # reset_index() : 데이트프레임 인덱스 재설정
ns_book4.head()
```
